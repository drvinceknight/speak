# Moran Process

```python
def moran(N, game, initial_population=None, seed=0):
    """
    Return the population counts for the Moran process on a game
    """   
    np.random.seed(seed)
    num_strategies = len(game)
    game = game - np.min(game) + 1
    
    if initial_population is None:
        population = np.random.randint(num_strategies, size=N)
    else:
        population = np.array([[i] * count 
                                   for i, count in enumerate(initial_population)])
        population = population.flatten()
        
    counts = [[(population == i).sum() for i in range(num_strategies)]]

    
    while len(set(population)) != 1:
        
        scores = []
        
        for i, player in enumerate(population):
            total = 0
            for j, opponent in enumerate(population):
                if i != j:
                    total += game[player, opponent]
            scores.append(total)

        total_score = sum(scores)
        probabilities = [score / total_score for score in scores]

        reproduce_index = np.random.choice(range(N), p=probabilities)
        
        eliminate_index = np.random.randint(N)
        population[eliminate_index] = population[reproduce_index]
        
        counts.append([(population == i).sum() for i in range(num_strategies)])
    return counts
```

```python
history = moran(10, rpsls, initial_population=[2, 2, 2, 2, 2])
plt.plot(history)
plt.legend(labels=labels);
```

# Demo for [pyconuk](http://www.pyconuk.org/)

```python
%matplotlib inline
from IPython.display import set_matplotlib_formats
from IPython.core.pylabtools import figsize
set_matplotlib_formats('svg')
figsize(8, 4)
import seaborn
import axelrod
```

```python
strategies = [axelrod.Cooperator(), axelrod.Defector(), axelrod.Grudger(), axelrod.TitForTat(), axelrod.Random()]
```

```python
trnmt = axelrod.Tournament(strategies)
```

```python
results = trnmt.play()
```

```python
plot = axelrod.Plot(results)
```

```python
plot.boxplot();
```

```python
evo = axelrod.Ecosystem(results)
evo.reproduce(100)
```

```python
plot.stackplot(evo.population_sizes);
```

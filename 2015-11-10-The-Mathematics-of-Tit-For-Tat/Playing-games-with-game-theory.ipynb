# Axelrod

Source code for some of the images and outputs for the slides.

```python
import axelrod
import seaborn
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
strategies = [axelrod.Defector(), axelrod.Alternator(), axelrod.Random(), axelrod.TitForTat(), axelrod.Cooperator()]
tournament = axelrod.Tournament(strategies)
results = tournament.play()
```

```python
strategies
```

```python

results.payoff_matrix
```

```python
plot = axelrod.Plot(results)
p = plot.boxplot();
p;
```

```python
eco = axelrod.Ecosystem(results)
eco.reproduce(100) # Evolve the population over 50 time steps
plot = axelrod.Plot(results)
p = plot.stackplot(eco) # Looks bad but svg is fine (I think it's a mpld3 thing)
p.savefig("static/axelrod.pdf");
```
